# ***Penyajian Data***

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv("https://raw.githubusercontent.com/yeftayudistira/data/main/Laptops.csv")
# Pembersihan data dan Konversi Mata uang
rupiah = 188.273
data['Price'] = data['Price'].str.replace('₹', '').str.replace(',', '').astype(int)*rupiah
data['Screen Size'] = data['Screen Size'].str.extract(r"([0-9]*\.?[0-9]+)").astype(float)
data
```

Penyajian Data jumlah laptop berdasarkan Brand

```python
brand_laptop = data.Brand.value_counts()
count_brand = brand_laptop.tolist()

brand_laptop.plot.bar(x='Brand', y ='count', title = 'Kategori Brand Laptop')

for i, v in enumerate(count_brand):
    plt.text(i, v + 0.5, str(v), ha='center')
```

Penyajian rata rata Harga untuk setiap Brand

```python
mean_price = data.groupby('Brand')['Price'].mean()
mean_price
```

```python
mean_price.plot.bar(title = 'Rata Rata Harga pada setiap Brand')
```

Pembagian Kategori Ram dengan diagram line

```python
ram_laptop = data.RAM.value_counts()
count_ram = ram_laptop.tolist()
ram_laptop
```

```python
ram_laptop.plot.line(title='Kategori Ram Laptop', linestyle = 'dotted')

for i, v in enumerate(count_ram):
    plt.text(i, v + 0.5, str(v), ha='center')
```

Penyajian Data untuk melihat Brand laptop terbesar berdasarkan data sebelumnya

```python
ram_brand_hp = data.loc[data['Brand'] == 'HP',['RAM']].value_counts()
ram_brand_hp.plot.pie(title = 'Kategori Ram pada brand laptop terbesar HP',autopct='%1.1f%%')
```

```python
processor_brand_hp = data.loc[data['Brand'] == 'HP',['Processor']].value_counts()
count_prohp = processor_brand_hp.tolist()

processor_brand_hp.plot.bar(x='Processor', y ='count', title = 'Kategori Proccesor Brand Laptop HP')

for i, v in enumerate(count_prohp):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')
```

Kategori Pada ukuran layar Laptop

```python
data['Screen Size'].hist(bins=10)
```

Mencari sistem operasi paling populer yang digunakan pada laptop

```python
system = data.groupby(['Operating System'])['Operating System'].count()
count_system = system.tolist()
system.plot.bar(title = 'Most Populer Operating System')
for i, v in enumerate(count_system):
    plt.text(i, v + 0.5, str(v), ha='center', va='bottom')
```

Rata Rata Harga Laptop berdasarkan Processor

```python
pro_price = data.groupby('Processor')['Price'].mean()
pro_price.plot.bar(title = 'Rata Rata Harga Laptop berdasarkan Processor')
```

**Desil**

```python
# Pembersihan data (jika diperlukan)
data_clean = data.dropna(subset=['Price'])

# Pembagian desil
n = [10, 20, 30, 40, 50, 60, 70, 80, 90]

# Fungsi untuk menghitung desil
desil = np.percentile(data_clean['Price'], n)

# Menampilkan hasil
for i, d in enumerate(desil, start=1):
    print(f"Desil {i} = {d}")
```

Ukuran Simpangan, dan Jarak Antar kuartile pada Price

```python
data_clean = data.dropna(subset=['Price'])

q3 = np.percentile(data_clean['Price'], 75)
q1 = np.percentile(data_clean['Price'], 25)
jarak_antar_kuartil = q3 - q1
print("Jarak Antar Kuartil:", jarak_antar_kuartil)

mean = np.mean(data_clean['Price'])
simpangan_baku = np.sqrt(np.sum((data_clean['Price'] - mean)**2) / len(data_clean))
print("Simpangan Baku:", simpangan_baku)

simpangan_kuartil = jarak_antar_kuartil / 2
print("Simpangan Kuartil:", simpangan_kuartil)
```
